```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import plot_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

# Set random seed for reproducibility
tf.random.set_seed(42)

# Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the data
x_train, x_test = x_train / 255.0, x_test / 255.0

# Further split train set into train and validation
x_train, x_val, y_train, y_val = train_test_split(
    x_train, y_train, test_size=0.2, random_state=42
)

# Function to build a model with given activation and dropout
def build_model(activation='relu', use_dropout=False):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(128, activation=activation))

    if use_dropout:
        model.add(Dropout(0.3))  # Dropout layer to prevent overfitting

    model.add(Dense(64, activation=activation))

    if use_dropout:
        model.add(Dropout(0.3))

    model.add(Dense(10, activation='softmax'))  # Output layer
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# Function to plot accuracy and loss curves
def plot_history(history, title='Model Performance'):
    plt.figure(figsize=(12, 5))

    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Train and evaluate models with ReLU and Sigmoid
activations = ['relu', 'sigmoid']
histories = {}

for act in activations:
    print(f"\nTraining model with activation: {act.upper()}")
    model = build_model(activation=act, use_dropout=False)
    history = model.fit(
        x_train, y_train,
        validation_data=(x_val, y_val),
        epochs=10,
        batch_size=128,
        verbose=1
    )
    histories[act] = history
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    print(f"Test Accuracy with {act.upper()}: {test_acc:.4f}")
    plot_history(history, title=f'{act.upper()} Activation')

# Model with Dropout (using best activation - typically ReLU)
print("\nTraining model with ReLU + Dropout Regularization")
dropout_model = build_model(activation='relu', use_dropout=True)
dropout_history = dropout_model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=10,
    batch_size=128,
    verbose=1
)

dropout_test_loss, dropout_test_acc = dropout_model.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy with Dropout: {dropout_test_acc:.4f}")
plot_history(dropout_history, title="RELU + Dropout")

# Show model architecture
print("\nModel Architecture:")
dropout_model.summary()

# Save model architecture image (optional)
try:
    os.makedirs("outputs", exist_ok=True)
    plot_model(dropout_model, to_file="outputs/model_architecture.png", show_shapes=True, show_layer_names=True)
    print("\nModel architecture image saved to outputs/model_architecture.png")
except:
    print("Could not save model architecture image (pydot not installed)")
```
